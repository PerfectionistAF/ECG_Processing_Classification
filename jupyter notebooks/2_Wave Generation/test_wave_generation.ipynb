```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
```

```python
path = 'ptb_ecg_db_csv/169.csv'  ##example vectorcardiogram fro vx,vy,vz
ex_df = pd.read_csv(path)
```

```python
ex_df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

ann = ['x','y','z']

x = ex_df['vx']
y = ex_df['vy']
z = ex_df['vz']

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z, color='blue', label='Original')

ax.set_xlabel('Vx')
ax.set_ylabel('Vy')
ax.set_zlabel('Vz')
ax.set_title('Vectorcardiogram of Healthy Patient')

ax.grid(True)
ax.legend()
ax.view_init(elev=20)
plt.show()
```

```python
path = 'ptb_ecg_db_csv/1.csv'  ##example vectorcardiogram fro vx,vy,vz
ex_df_1 = pd.read_csv(path)
x = ex_df_1['vx']
y = ex_df_1['vy']
z = ex_df_1['vz']

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z, color='red', label='Original')

ax.set_xlabel('Vx')
ax.set_ylabel('Vy')
ax.set_zlabel('Vz')
ax.set_title('Vectorcardiogram of Sick Patient')

ax.grid(True)
ax.legend()
ax.view_init(elev=20)
plt.show()
```
